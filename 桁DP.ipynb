```python
N = '12345'
length = len(N)
dp = [[0]*2 for _ in range(length+1)]
dp[0][0] += 1

for i in range(length):  ## いま数え上げている桁
    max_digit = int(N[i])

    # flag_less: 過去に対応する桁より小さい数を入れたことがある時はTrue=1
    # Trueの時、0から9の数字を使える ex. 12345 の2桁目まで 11*** であれば、
    # ３桁目はどの数字でも良い。
    # Falseの時、その桁の数字が使える最大の数字 ex. 12345 2の2桁目まで 12***であれば、
    # 3桁目は 0 or 1 or 2 or 3 のいずれか。
    for flag_less in range(2):
        range_digit = 9 if flag_less else max_digit
        for d in range(range_digit + 1): ## いま数え上げている桁を実際数え上げる
            flag_less_next = 0
            ### flag_less==0かつd=max_digitのとき
            ### つまり過去に小さい数を入れておらず
            ### 数えてる数が桁と一致するとき０
            if flag_less==1 or d<max_digit:
                flag_less_next = 1
            dp[i+1][flag_less_next] += dp[i][flag_less]
print(dp)
print(dp[length][0] + dp[length][1])
```

```python
############ 結局数え上げている事実は変わらないのである。
########### それを桁ごとにやっているだけ
N = '12345'
length = len(N)

# dp[決めた桁数][未満フラグ][4or9が入っている] := 総数
dp = [[[0]*2 for _ in range(2)] for _ in range(length+1)]
dp[0][0][0] = 1

for i in range(length):
    max_digit = int(N[i])
    for flag_less in range(2):
        for flag_four_or_nine in range(2):
            range_digit = 9 if flag_less else max_digit
            for d in range(range_digit+1):
                flag_less_next = 0
                flag_four_or_nine_next = 0
                if flag_less==1 or d<max_digit:
                    flag_less_next = 1
                if flag_four_or_nine==1 or d==4 or d==9:
                    flag_four_or_nine_next = 1
                dp[i+1][flag_less_next][flag_four_or_nine_next] += \
                            dp[i][flag_less][flag_four_or_nine]
print(dp[length][0][1] + dp[length][1][1])
```

```python
## dp[決めた桁数][未満フラグ][3が入っている][5が入っている][7が入っている][3,5,7以外が入っている]
import numpy as np
N = input()
length = len(N)

dp = np.zeros((length+1, 2, 2, 2, 2,2))
dp[0,0,0,0,0,0] = 1

for i in range(length):
    max_digit = int(N[i])
    
    for flag_less in range(2):
        for three in range(2):
            for five in range(2):
                for seven in range(2):
                    for others in range(2):
                    
                        range_digit = 9 if flag_less else max_digit
                        for d in range(range_digit + 1):

                            flag_less_next = 0
                            three_next = 0
                            five_next  = 0
                            seven_next = 0
                            others_next = 0

                            if flag_less==1 or d<max_digit:
                                flag_less_next = 1
                            if three==1 or d==3:
                                three_next = 1
                            if five==1 or d==5:
                                five_next = 1
                            if seven==1 or d==7:
                                seven_next = 1
                            if others==1 or (d in [0,1,2,4,6,8,9]):
                                others_next = 1

                            dp[i+1][flag_less_next][three_next][five_next][seven_next][others_next] \
                                 += dp[i][flag_less][three][five][seven][others]
print(int(dp[length, 0, 1, 1, 1, 0] + dp[length, 1, 1, 1, 1, 0]))   
```
